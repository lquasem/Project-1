```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import csv

# Read csv file
restaurants_df = pd.read_csv("Resources/restaurant_data_coords2.csv", encoding = "ISO-8859-1", 
                             names=["search_ID", "Latitude_search", "Longitude_search", "Name", "Lat_Restaurant", "Lng_Restaurant", "Address", "Zip", 
                                   "Rating", "# of Reviews", "Price Level", "Category", "Yelp ID"],
                             low_memory=False)

#preview the dataframe
restaurants_df.head()
```

```python
#groupby coordinate search, to get counts of restaurants in the area
group_df = restaurants_df.groupby("search_ID")
counts_by_coord = group_df["search_ID"].count().to_frame()

counts_by_coord
```
