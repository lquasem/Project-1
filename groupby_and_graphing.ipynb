# READ CSV FOR RANDOM TESTS & ANALYSIS

For easy access to csv files to allow quick manipulating and testing of data

```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import csv

# Read csv file
clean_data = pd.read_csv("Merged_Zip.csv", encoding = "ISO-8859-1",
                             low_memory=False)

#preview the dataframe
clean_data = clean_data.drop('Unnamed: 0', axis=1)

clean_data
```

```python
clean_data_high_ratings = clean_data[clean_data["Rating"]>=3.5]
clean_data_high_ratings
```

```python
clean_data_high_ratings["Rating"].unique()
```

```python
clean_data_high_ratings["Rating"].value_counts()
```

```python
high_ratings_grouby_county = clean_data_high_ratings.groupby("County")
high_ratings_grouby_county
```

```python
#break down of the counts of the different categories with high ratings by county
category_counts = high_ratings_grouby_county["Category"].value_counts().to_frame()
category_counts.columns.values[0] = "Counts"
category_counts.reset_index().head(20)
```

```python
#number of categories  
number_of_categories = high_ratings_grouby_county["Category"].nunique().to_frame()
number_of_categories
```

```python
number_of_categories.sort_values("Category", ascending=False)
```

```python
#group by county with all restaurant ratings
groupby_2 = clean_data.groupby(["County"])
groupby_2
```

```python
#find out the average rating for by county, with standard deviation
average_rating = groupby_2["Rating"].mean()
average_rating
```
