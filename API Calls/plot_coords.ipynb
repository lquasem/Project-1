```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np
import requests
import time
import json
import csv
```

```python
coords1 = pd.read_csv("Resources/total_count_coords.csv")
```

```python
plt.scatter(coords1["Lng"], coords1["Lat"], marker="o", s=coords1["Total # of Restaurants"],
            facecolors="red", alpha = 0.3, edgecolors="black")

plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.grid()
plt.figure().savefig('coords_map.pdf')
plt.show()
```

```python
# Code from stackoverflow to adjust ratio, also the comments

plt.style.use('ggplot')
x = np.linspace(-5, 5, 100)
y1 = np.exp(0.8*x)
y2 = np.sin(x)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(coords1["Lng"], coords1["Lat"], marker="o", s=coords1["Total # of Restaurants"],
           facecolors="red", alpha = 0.3, edgecolors="black")


ratio = 1.5
xleft, xright = ax.get_xlim()
ybottom, ytop = ax.get_ylim()
# the abs method is used to make sure that all numbers are positive
# because x and y axis of an axes maybe inversed.
ax.set_aspect(abs((xright-xleft)/(ybottom-ytop))*ratio)
# or we can utilise the get_data_ratio method which is more concise
# ax.set_aspect(1.0/ax.get_data_ratio()*ratio)


plt.savefig("Resources/coords_plot.png", bbox_inches='tight')
```
