```python
county_zips = pd.read_csv("Resources/Zip of 9 counties.csv", header=0)

del county_zips['Unnamed: 0']
del county_zips['Unnamed: 1']
del county_zips['Unnamed: 2']


county_zips_clean = county_zips.dropna(how='all')

county_zips_clean = county_zips_clean.astype("object")




county_zips_clean.to_csv("Cleaned_zip.csv")

county_zips_clean
```

```python
a=county_zips_clean['Napa'].unique()
b =county_zips_clean['SF'].unique()
c=county_zips_clean['Marin'].unique()
d=county_zips_clean['Sonoma'].unique()
e =county_zips_clean['Contra Costa'].unique()
f=county_zips_clean['Alameda'].unique()
g=county_zips_clean['San Mateo'].unique()
h =county_zips_clean['Santa Clara'].unique()
i =county_zips_clean['Solano'].unique()

x = np.concatenate((a,b,c,d,e,f,g,h,i),axis=0)


x
```

```python
y = []

for zip in x:
    if zip > 0:
        y.append(str(int(zip)))

# clean list of good zip codes
y
```

```python
for i in range(0,len(restaurants_clean_1)):
    print(f'{len(restaurants_clean_1)-i} datapoints remaining')
    biz = restaurants_clean_1.iloc[i]
    storage = [biz["Name"], biz["Category"], biz["Biz City"], biz["Biz Zip"], biz["Biz Lat"], 
               biz["Biz Lng"], biz["Rating"], biz["Review Count"], biz["Price"], biz["Yelp ID"]]
    if pd.isna(biz["Biz Zip"]):
        with open('Resources/unclean_2.csv', 'a', encoding="utf-8") as csvFile:
            writer = csv.writer(csvFile)
            writer.writerows([storage])
        csvFile.close()
    else:
        try:
            biz_zip = str(int(float(biz["Biz Zip"])))
        except ValueError:
            biz_zip = ''
        if biz_zip in y:
            with open('Resources/cleaning_2.csv', 'a', encoding="utf-8") as csvFile:
                writer = csv.writer(csvFile)
                writer.writerows([storage])
            csvFile.close()
        else:
            with open('Resources/unclean_2.csv', 'a', encoding="utf-8") as csvFile:
                writer = csv.writer(csvFile)
                writer.writerows([storage])
            csvFile.close()
        
# if Zipcode in county zips then keep it
```
